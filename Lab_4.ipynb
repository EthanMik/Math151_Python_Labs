```python
import sympy as sp
from sympy import *
from sympy.plotting import (plot,plot_parametric, plot_implicit)
```

### Question 1

#### 1a

```python
x = symbols("x")

f = x**4 - 50 * x * x + 15 * x + 635

df = diff(f, x)
tangent_x_values = solve(df - 15, x)

print(f"Line is tangent at x = {str(tangent_x_values).strip('[]')}")
```

#### 1b

```python
f = x**4 - 50 * x * x + 15 * x + 635

tangent_x = tangent_x_values[0]
tangent_y = f.subs(x, tangent_x)
tangent_slope = 15

tangent_line = (tangent_slope * (x - tangent_x)) + tangent_y 

p1 = plot(f, (x, -6, 6), show=False)
p2 = plot(tangent_line, (x, -6, 6), show=False)
p1.extend(p2)
p1.show()
```

#### 1c

```python
x = symbols("x")
y = symbols("y")

f = (144 * (x**4 + y**4)) - (255 * (x**2 + y**2)) + (350 * x**2 * y**2) + 81

plot_implicit(f, (x, -1, 1), (y, -1, 1))
```
