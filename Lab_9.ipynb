```python
ddf = 8 / (x + 1)**3

df = integrate(ddf, x) + C
c = solve(df.subs(x, 0) - 3, C)[0]
df = df.subs(C, c)
print(f"f'(x) = {df}")

f = integrate(df, x) + C
c = solve(f.subs(x, 0) - 9, C)[0]
f = f.subs(C, c)
print(f"f(x) = {f}")
```

#### 3b

```python
df = integrate(ddf, x) + C
f = integrate(df, x) + D

display(f)

f1 = f.subs(x, 1)
f2 = f.subs(x, 7)

solution = nsolve((f1, f2), (C, D), (1, 1))
print(solution)

display(f.subs({C : solution[0], D : solution[1]}))
```
