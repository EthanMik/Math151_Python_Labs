### Question 1

```python
x_t = (E ** (-0.2 * t)) * cos(2 * t)
y_t = (E ** (-0.2 * t)) * sin(2 * t)

dydt = diff(y_t, t)
dxdt = diff(x_t, t)
dydx = dydt / dxdt

display(dydx)
```

### Question 2

#### 2a

```python
eq = y0 * exp(r * t) - y

eq1 = eq.subs({y:15000, t:1})
eq2 = eq.subs({y:19000, t:5})

solution = solve((eq1, eq2), (y0, r))[0]
print(f"y0 = {solution[0].evalf(5)}")
print(f"r = {solution[1].evalf(5)}")
```

#### 2b

```python
model = eq.subs({y:21000})
solution = solve(model, t)[0]
print(f"The account will reach $21,000 in {solution.subs({y0:14139, r:0.05910}):.2f} years")
```

#### 2c

```python
model = eq.subs({y0:15000, y:19000, t:4})
r1 = solve(model, r)[0].evalf(5)

value_eq = eq.subs({y0:15000, r:r1, t:-1})
value = solve(value_eq, y)[0].evalf(5)

print(f"The new value of r is {r1}")
print(f"The value of the account 1 year before intial time = ${value}")
print(f"The y at t = -1, is the same y0 value calculated in 2a")
```
