```python
A, B = symbols("A B", real=True)

ellip_curve_form = x**3 - A * x + B - y**2
display(ellip_curve_form)

ellip_curve = x**3 - 3 * x + 2 - y**2
display(ellip_curve)

singular_eq = 4 * A**3 + 27 * B**2
display(singular_eq)

print("4(-3)^3 + 27(2)^2 =", singular_eq.subs({A: -3, B: 2}))
print("The curve is singular")

plot_implicit(ellip_curve, (x, -5, 5), (y, -5, 5))
```

### Question 2

#### 2a

```python
dy = symbols("d")

f = dy - 2 * (ln(2 * x) + 1) * y

Y = (2 * x)**(2 * x)
dY = diff(Y, x)

result = f.subs({y: Y, dy: dY})

print(f"When y and y' are subbed into the equation, the result is \"{result}\", \nTherefore, y = (2x)^2x is a solution of the differential equation")
```

#### 2b

```python
a, k = symbols('a k', real = True)

f = dy - k * (ln(k * x) + 1) * y

Y = a * (k * x)**(k * x)
dY = diff(Y, x)

result = simplify(f.subs({y: Y, dy: dY}))

print(f"When y and y' are subbed into the equation, the result is \"{result}\", \nTherefore, y = a(kx)^kx is a solution of the differential equation")
```

```python
a, c = symbols("a, c", real=True)

f = (x**2 + y**2 + a**2)**2 - (4 * a**2 * x**2) - c**4

dydx = idiff(f, y, x)

display(dydx)
```

#### 3b

```python
d2ydx = idiff(dydx.subs(a, 1), y, x)

display(d2ydx)
```

#### 3c

```python
f = f.subs({a: 1, c:2})

num = x * ((-x**2 / y) - y + (1 / y))
den = x*x + y*y + 1

htan = solve([num, f], [x, y])
print(f"Horizontal tangents at {htan}")

vtan = solve([den, f], [x, y])
print(f"Vertical tangent at {vtan}")

```
